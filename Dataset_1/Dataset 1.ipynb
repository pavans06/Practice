```python
df = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv',sep='\t')
```

```python
l = len(df['item_price'])  #no of item_price in the series
```

```python
for i in range(l):
    v = df['item_price'][i]                       # getting individual row values inside column item_price
    res = v[1:]                                   # removing the $ sign infront of the string
    df['item_price'][i] = float(res)              # coverting to float  
```

```python
df[['item_price']]= pd.to_numeric(df['item_price'])
```

```python
total_price = df['item_price'].sum()  #total items price for all orders
```
